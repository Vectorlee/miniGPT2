```python
import matplotlib.pyplot as plt
```

```python
log_file = "fineweb_edu_10B/log_file.txt"
```

```python
train_loss = []
val_loss = []
```

```python
with open(log_file) as pfile:
    for line in pfile:
        parts = line.strip().split(' ')
        if parts[1] == 'train':
            train_loss.append((int(parts[0]), float(parts[2])))
        else:
            val_loss.append((int(parts[0]), float(parts[2])))
```

```python
def moving_average(data, window_size=20):
    """Compute moving average with a sliding window."""
    return [sum(data[i:i+window_size]) / window_size for i in range(len(data) - window_size + 1)]
```

```python
# Unpack tuples into x and y for each list
#x1, y1 = zip(*train_loss)  # unzip list1 into x1 and y1
#x2, y2 = zip(*val_loss)  # unzip list2 into x2 and y2

plt.subplots(figsize=(12, 5))

loss_list = list(map(lambda s: s[1], train_loss))
loss_list = moving_average(loss_list)

# Plot both lines
plt.plot(loss_list, color='blue', label='Train Loss')
#plt.plot(x2, y2, color='green', label='Val Loss')

# Labels, title, legend
plt.xlabel('Steps')
plt.ylabel('Loss')
#plt.xscale('log')
plt.legend()

# Show plot
#plt.show()
```
